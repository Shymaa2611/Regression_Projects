```python
theta0=1.0
theta1=2.0
theta2=1
theta3=1
alpha=0.1
```

```python

def get_theta0_theta1(theta0,theta1,theta2,theta3):
    sumvalue=0
    m=len(data)
    for i in range(1,1000):
        for j in range(1,m+1):
            hTheta=theta0+theta1*x[j-1][0]+theta2*x[j-1][1]+theta3*x[j-1][2]
            sumvalue=sumvalue+(hTheta-y[j-1])
        theta0=theta0-alpha*(1/m)*sumvalue
        theta1=theta1-alpha*(1/m)*sumvalue
        theta2=theta2-alpha*(1/m)*sumvalue
        theta3=theta3-alpha*(1/m)*sumvalue
    return theta0,theta1,theta2,theta3
    
```

```python
theta0,theta1,theta2,theta3=get_theta0_theta1(theta0,theta1,theta2,theta3)
```

```python
def prediction(theta0,theta1,theta2,theta3):
    y_predict=[]
    m=len(data)
    for j in range(m):
            hTheta=theta0+theta1*x[j][0]+theta2*x[j][1]+theta3*x[j][2]
            y_predict.append(hTheta)
    return y_predict
```

```python
y_predict=prediction(theta0,theta1,theta2,theta3)
```

```python
def calculate_cost_function():
    m=len(data)
    for i in range(m):
        sumValue=(y_predict[i]-y[i])**2
    cost_function=(1/(2*m))/sumValue
    return cost_function
```
